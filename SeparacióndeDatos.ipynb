## Verificación de que el dataset este completo y listo para trabajar.

```python
# Supongamos que tienes un DataFrame llamado "df" con una columna de fechas llamada "Fecha de contabilización".
# Asegúrate de que la columna de fecha esté en formato datetime.

# Definir las fechas de inicio y fin para cada conjunto
fecha_inicio_entrenamiento = "2020-02-01"
fecha_fin_entrenamiento = "2022-12-31"
fecha_inicio_validacion = "2023-01-01"
fecha_fin_validacion = "2023-06-30"
fecha_inicio_prueba = "2023-06-01"
fecha_fin_prueba = "2023-08-31"

# Filtrar los datos en los conjuntos de entrenamiento, validación y prueba
conjunto_entrenamiento = df[(df['Fecha de contabilización'] >= fecha_inicio_entrenamiento) & (df['Fecha de contabilización'] <= fecha_fin_entrenamiento)]
conjunto_validacion = df[(df['Fecha de contabilización'] >= fecha_inicio_validacion) & (df['Fecha de contabilización'] <= fecha_fin_validacion)]
conjunto_prueba = df[(df['Fecha de contabilización'] >= fecha_inicio_prueba) & (df['Fecha de contabilización'] <= fecha_fin_prueba)]
```

## En el anterior codigo se realiza la separación de datos mediante 3 conjuntos, conjunto de entrenamiento para los dos primeros años, conjunto de validación para los primeros 6 meses del 2023 y el ultimo conjunto de prueba para los meses de Junio, Julio y Agosto de 2023.
